```python
import numpy as np
import os
os.environ['OMP_NUM_THREADS'] = '1'
from sklearn.cluster import KMeans
data = np.loadtxt('4d clusteringdataset.csv', delimiter=',', skiprows=1)
kmeans = KMeans(n_clusters=8, n_init=10)
kmeans.fit(data)
labels = kmeans.predict(data)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
n_samples = 350
n_features = 2
n_clusters = 3
X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
n_samples = 350
n_features = 2
n_clusters = 3

X, y = make_blobs(n_samples=n_samples, n_features=n_features, centers=n_clusters, random_state=42)
kmeans = KMeans(n_clusters=n_clusters, init='random', n_init=10, random_state=42)
kmeans.fit(X)
cluster_centers = kmeans.cluster_centers_
cluster_labels = kmeans.labels_
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cluster_labels, cmap='viridis', edgecolor='k', label='Data Points')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='x', s=100, label='Cluster Centers')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar()
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(data, labels)
print("Silhouette Score:", silhouette_avg)
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(0)
X = np.random.rand(100, 4) 
kmeans.fit(X)
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
plt.figure(figsize=(8, 8))
sns.heatmap(cluster_centers, annot=True, cmap='YlGnBu', fmt='.4f', cbar=True, square=True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('K-Means Cluster Centers Heatmap')
plt.show()
```
